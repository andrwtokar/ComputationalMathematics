## В следующей ячейке создадим матрицу М, которая является матрицей  СЛАУ, а также введем столбец свободных членов.

```
# Создаем матрицу СЛАУ.
M = np.array(100*[np.zeros(100)])

print("\nТаблица из \'0\' размера 100х100.\n")
print(M)
print("\n\nЗаполненная таблица!\n\n")

for k in range(100):
    M[k][k] = 20;
    if k < 4:
        for i in range(k):
            M[k][i] = 1
        for i in range(k + 1, k + 5):
            M[k][i] = 1

    if k > 3 and k < 96:
        M[k][k - 4] = 1
        M[k][k - 3] = 1
        M[k][k - 2] = 1
        M[k][k - 1] = 1
        M[k][k + 1] = 1
        M[k][k + 2] = 1
        M[k][k + 3] = 1
        M[k][k + 4] = 1
    
    if k > 95:
        for i in range(k + 1, 100):
            M[k][i] = 1
        for i in range(k - 4, k):
            M[k][i] = 1


print(M)

# Далее кусок кода, которым можно полностью вывести матрицу,
# но в ней сложно ориентироваться тогда.
#
# for row in M:
#     print(' '.join(str(col) for col in row))

# Создаем столбец свободных членов.
f = np.zeros(100)
for i in range(100):
    f[i] = i + 1

print("\n\nНиже представлен столбец свободных членов.\n\n {}".format(f))
```

Ниже напишем две функции, которые высчитывают третьи (спектральные) нормы вектора и матрицы.

```
def NormaVector(x):
    sum = 0
    for i in x:
        sum += i**2
    return math.sqrt(sum)

def NormaMatrix(A):
    lamb, vect = np.linalg.eig(A.dot(A.T))
    max = -1000
    for i in lamb:
        if i > max:
            max = i
    return math.sqrt(math.abs(max))
```

# Метод Гаусса
Ниже рассмотрим решение СЛАУ методом Гаусса. 
- пройдемся по расширенной матрице прямым ходом метода Гаусса
- потом пройдемся обратным ходом и получим в итоге диагональную матрицу СЛАУ и некоторый столбец свободных членов
- получим ответ для данного метода
- выведем на печать невязки данного способа

```
A = np.copy(M)
f1 = np.copy(f)
for k in range(100):
    a = A[k][k]
    if (99 - k) > 3:
        for i in range(1, 5):
            f1[k + i] = f1[k + i] - f1[k]*(A[k + i][k])/a
            A[k + i] = A[k + i] - A[k]*(A[k + i][k])/a
    else:
        for i in range(1, 99 - k + 1):
            f1[k + i] = f1[k + i] - f1[k]*(A[k + i][k])/a
            A[k + i] = A[k + i] - A[k]*A[k + i][k]/a
    for i in range(100):
        if A[k][i] < 10**(-15):
            A[k][i] = 0


l = list(range(100)[::-1])
for k in l:
    a = A[k][k]
    if k > 3:
        for i in range(1, 5):
            f1[k - i] = f1[k - i] - f1[k]*(A[k - i][k])/a
            A[k - i] = A[k - i] - A[k]*(A[k - i][k])/a
    else:
        for i in range(1, k + 1):
            f1[k - i] = f1[k - i] - f1[k]*(A[k - i][k])/a
            A[k - i] = A[k - i] - A[k]*(A[k - i][k])/a
    for i in range(100):
        if A[k][i] < 10**(-15):
            A[k][i] = 0

print("Вывод столбца свободных членов.\n")
print(f1)

# Далее кусок кода, которым можно полностью вывести матрицу,
# но в ней сложно ориентироваться тогда.
#
# for row in A:
#     print(' '.join(str(col) for col in row))

print("\nВывод преобразованной с помощью прямого и обратного ходов Гаусса матрицы СЛАУ.\n")
print(A)

res = np.zeros(100)

for i in range(len(res)):
    res[i] = f1[i]/(A[i][i])

print("\nВывод получившегося решения.\n")
print(res)

A = np.copy(M)
r = A.dot(res) - f
print("\nВывод невязок.\n")
print(r)

for i in range(len(r)):
    if r[i] < 10**(-15):
        r[i] = 0

print("\nВывод на печать невязок, при условии, что все числа меньше е-16 приравняны 0.\n\n {}".format(r))
print("\nНорма невязок методом Гаусса: {}".format(NormaVector(A.dot(res) - f)))
```

# Собственные числа
Ниже рассмотрим все собственные значения нашей матрицы и выведем минимальное и максимальное из них.

```
# Функция, для расчета и вывода собственных значений матрицы, 
# а также вывода максимального и минимального собственных значений.

# Возвращает вектор собственных значений, а также соответствующие им собственные вектора.

def EigenValues(A):
    l, v = np.linalg.eig(A)
    print(l)

    minValue = 10**6
    maxValue = -1000
    for i in l:
        if i < minValue:
            minValue = i
        if i > maxValue:
            maxValue = i

    print("\nMin lambda: {:>3}.\nMax lambda: {:>3}.".format(minValue, maxValue))

    return (l, v)
```

```
# lam, vec = np.linalg.eig(M)
print("Ниже представлены все собстенные значения данной нам матрицы.\n")
lam, vec = EigenValues(M)
```

```
m = np.linalg.inv(M)

print("Ниже представлены все собвственные числа обратной матрицы.\n")
lam1, vec1 = EigenValues(m)
```

```
# Рассчитаем число обусловленности по 3 норме.

u = NormaMatrix(m) * NormaMatrix(M)
print("Число обусловленности, путем создания функции, вычисляющей норму матрицы: u = {}.".format(u))
print("\nФункция для определения норм матрицы и вектора приведены в начале документа.")
```

# Метод Зейделя
Делать метод Зейделя в лоб кажется мне очень сложным делом. Поэтому я применю разложение матрицы на верхне-, нижне- треугольные и диагональную.

```
A = np.copy(M)
f1 = np.copy(f)
L = np.array(100*[np.zeros(100)])
D = np.array(100*[np.zeros(100)])
V = np.array(100*[np.zeros(100)])

for i in range(100):
    D[i][i] = 20
    for j in range(100):
        if i < j and i > j - 5:
            V[i][j] = 1
        if i > j and i < j + 5:
            L[i][j] = 1
print("Нижнетреугольная матрица.\n")
print(L)
print("\nДиагональная матрица.\n")
print(D)
print("\nВерхнетреугольная матрица.\n")
print(V)
```
